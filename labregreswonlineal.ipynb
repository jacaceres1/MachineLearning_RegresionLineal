# Importado de Librerias

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
from sklearn import linear_model
```

# Generación de los datos



```python

"""
        Receives by parameter the url of the csv file
        df = pandas.read_csv(url)
        Print the five datas of the file using the variable df 
"""
url = 'C:/Users/JoSe/Documents/Sep/Mineria/labs/LabLinear/insurance.csv'
df = pd.read_csv(url)
df.head(5)

    
```

# Obtener los valores Objetivos y los valores de Entrada

```python
"""
    features : genera la obtención de la información del data set en la edad
    labels: genera la obtención de la información del data set en el precio de seguro
    x: search in the dataset by storing in the variable of x
    y: searches the dataset by storing in the variable of y
    print: imprimen cada variable generada
"""
features = ['age']
labels = ['charges']
X = df[features].values
y = df[labels].values
print("Valores de y: \n")
print(y,"\n")

print("Valores de X: \n")
print(X)
```

calcula la regresion lineal de minimos cuadrados

# Separar las varibables

```python
"""
    X_train, Y_train : implements the transfer of data for training
"""
X_train = np.array(X)
y_train = y
```

# Regresion Lineal 

```python
"""
   regr: implementa la regresión lineal de mínimos cuadrados ordinarios
"""
regr = linear_model.LinearRegression()
```

# Entrenar el modelo

```python
"""
    regr.fit(): trains the model by sending the corresponding values per parameter
"""
regr.fit(X_train, y_train)
```

# Predicción del modelo

```python
"""
    y_pred = regr.predict(): performs label prediction 
"""
y_pred = regr.predict(X_train)
print(y_pred)
```

visualizacion del modelo

```python
"""
    plt.plot():creates a 2D line plot of Y data versus the corresponding X and Y values
    plt.legend(): A legend is an area that describes the elements of the chart.
"""
plt.plot(X,y,'*',label = 'original data')
plt.plot(X,y_pred,'r',label='fitted line')
plt.legend()
```
